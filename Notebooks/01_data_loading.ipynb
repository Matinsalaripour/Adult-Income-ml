# Import dataset

## Libraries

```python
import pandas as pd
from sklearn.datasets import fetch_openml
import os
```

## Import Dataset

```python
adult = fetch_openml(name='adult', version=2, as_frame=True)
df = adult.frame
```

## Save to raw Folder

```python
os.makedirs("../data/raw", exist_ok=True)
df.to_csv("../data/raw/adult.csv", index=False)
```
